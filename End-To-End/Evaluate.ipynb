```python
# https://huggingface.co/docs/datasets/installation
from datasets import load_metric
from transformers import T5Tokenizer
import pandas as pd
import warnings
import os
warnings.filterwarnings('ignore')
```

```python
to_eval = pd.read_csv('preliminary_results_length.csv',index_col=0)
ground_truth = to_eval['de'].to_list()
to_eval_pretrained = to_eval['de_translation_t5'].to_list()
to_eval_finetuned = to_eval['de_translation_t5_trained'].to_list()
```

```python
to_eval
```

```python
to_eval_pretrained_meteor = [s.replace('<pad> ','').replace("</s>",'').replace(" <pad>",'') for s in to_eval_pretrained ]
to_eval_finetuned_meteor = [s.replace('<pad> ','').replace("</s>",'').replace(" <pad>",'') for s in to_eval_finetuned ]
ground_truth_meteor = [s for s in ground_truth]
predictions = to_eval_pretrained
src = to_eval['en'].to_list()
references = ground_truth
```

```python
to_eval_pretrained_bleu = [s.replace('<pad> ','').replace("</s>",'').replace(" <pad>",'').split(" ") for s in to_eval_pretrained ]
to_eval_finetuned_bleu = [s.replace('<pad> ','').replace("</s>",'').replace(" <pad>",'').split(" ") for s in to_eval_finetuned ]
ground_truth_bleu = [[s.split(" ")] for s in ground_truth]
```

```python
pretrained_bleu_score = bleu_metric.compute(predictions=to_eval_pretrained_bleu, references=ground_truth_bleu)
finetuned_bleu_score = bleu_metric.compute(predictions=to_eval_finetuned_bleu, references=ground_truth_bleu)
print("pretrained", pretrained_bleu_score['bleu'])
print("finetuned", finetuned_bleu_score['bleu'])
```

```python
pretrained_meteor_score = meteor_metric.compute(predictions=to_eval_pretrained_meteor, references=ground_truth_meteor)
finetuned_meteor_score = meteor_metric.compute(predictions=to_eval_finetuned_meteor, references=ground_truth_meteor)
print("pretrained", pretrained_meteor_score['meteor'])
print("finetuned", finetuned_meteor_score['meteor'])
```

```python
results = comet_metric.compute(predictions=to_eval_pretrained_meteor, references=references, sources=src)
```

```python
results
```

```python
import numpy as np
res_num = [v for v in results["scores"]]
res_array = np.array(res_num)
print(res_array.mean())
print(res_array.std())
```
