```python
import matplotlib.pyplot as plt
plt.figure(figsize=(12, 6))
import pandas as pd
import numpy as np

figsize=(12, 6)
```

```python
def plot_graph(title, filename, legend):
    existing_task = pd.read_csv("outputs_training/existing_task/" + filename + ".csv", sep="=").iloc[:,2]
    plt.plot(existing_task, label=legend[0])
    
    existing_task_with_suffix = pd.read_csv("outputs_training/existing_task_with_suffix/" + filename + ".csv", sep="=").iloc[:,2]
    plt.plot(existing_task_with_suffix, label=legend[1])
    
    idiom_task = pd.read_csv("outputs_training/idiom_task/" + filename + ".csv", sep="=").iloc[:,2]
    plt.plot(idiom_task, label=legend[2])
    
    idiom_task_with_suffix = pd.read_csv("outputs_training/idiom_task_with_suffix/" + filename + ".csv", sep="=").iloc[:,2]
    plt.plot(idiom_task_with_suffix, label=legend[3])

    plt.legend()

    plt.title(title)

    plt.xlabel("Epochs")
    plt.ylabel(existing_task.name)
    plt.show()
    
```

```python
labels = ['Finetuned task', 'Finetuned task with idiom suffix', 'New task', 'New task with idiom suffix']
```

## Training loss

```python
plot_graph('Training loss', 'train_loss', labels)
```

## Validation loss

```python
plot_graph('Validation loss', 'val_loss', labels)
```

## Meteor score

```python
plot_graph('Meteor score', 'meteor', labels)
```
